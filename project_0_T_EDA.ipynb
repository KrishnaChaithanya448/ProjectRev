```python
#importing json
import json as js
#importing pandas
import pandas as pd
#importing matplotl
import matplotlib.pyplot as plt
#importing seaborn
import seaborn as sns
td = pd.read_json("transaction_logs.json")
print(td)
```

```python
############################################ EDA (Exploratory Data Analysis) #########################################

# 1.shape of the data (shape)
# it gives the how many number of rows and columns presented in data
td.shape 
```

```python
# 2.finding the only columns in the data (columns.tolist())
# it will gives the list of columns
td.columns.tolist()
```

```python
# 3.finding the nulls in data (isnull().sum())
# it will gives the sum of nulls in columns
td.isnull().sum()
```

```python
# 4.Info of the data (info())
# it gives the information about the data
td.info()
```

```python
# 5.finding the how many number of unique rows in each column (nunique())
# it will gives the count of unique rows in the each column 
td.nunique()
```

```python
# 6.finding the top or head rows in a data (head(n))
#it will returns the first n number of rows in the data
td.head(8)
```

```python
# 7.finding the bottom or tail rows in a data (tail(n))
# it will returns the last n number of rows in the data
td.tail(8)
```

```python
# 8.describing the data
td.describe()
```

```python
# Assuming 'df' is your DataFrame
quality_counts = td['qty'].value_counts()

# Using Matplotlib to create a count plot
plt.figure(figsize=(8, 6))
plt.bar(quality_counts.index, quality_counts, color='green')
plt.title('orders quantity')
plt.xlabel('quantity')
plt.ylabel('count')
plt.show()
```

```python
# Assuming 'df' is your DataFrame
quality_counts = td['payment_txn_success'].value_counts()

colors = ['lightgreen' if index == 'Y' else 'pink' for index in quality_counts.index]

# Using Matplotlib to create a count plot
plt.figure(figsize=(2, 3))
plt.bar(quality_counts.index, quality_counts, color=colors)
plt.title('payment_txn_success')
plt.xlabel('payment_txn_success')
plt.ylabel('count')
plt.show()
```

```python
# Assuming 'df' is your DataFrame
quality_counts = td['failure_reason'].value_counts()
colors = ['lightgreen' if index == '' else 'pink' for index in quality_counts.index]
# Using Matplotlib to create a count plot
plt.figure(figsize=(10, 5))
plt.bar(quality_counts.index, quality_counts, color=colors)
plt.title('failure_reason')
plt.xlabel('failure_reason')
plt.ylabel('count')
plt.show()
```
